Problem Statement:

* Plot Graph between Year and GDP value for given dataset
* This is a non- linear regression task as it involves just 2 variables.

```python
#import the main libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
#import the data file
GDP_data = pd.read_csv("china_gdp.csv")
```

```python
#plot the data
plt.figure(figsize=(20,10))
plt.plot("Year","Value","bo",data=GDP_data)
plt.xlabel("Year")
plt.ylabel("Value")
plt.title("China_GDP")
plt.grid()
plt.legend()
plt.show()
```

```python
# plot the bar chart
plt.figure(figsize=(20,10))
plt.bar(x="Year",height="Value",data=GDP_data)
plt.xlabel("Year")
plt.ylabel("GDP_Value")
plt.title("Chine_GDP")
plt.grid()
plt.show()
```

```python
# according to the graph. Data trend is non linear
x=GDP_data["Year"].values
y=GDP_data["Value"].values
```

```python
def sigmoid(x,Beta_1,Beta_2):
    y=1/(1+np.exp(-Beta_1*(x-Beta_2)))
    return y
Beta_1 = 0.10
Beta_2 = 1990.0
#logistic function
y_pred=sigmoid(x,Beta_1,Beta_2)
```

```python
#plot initial prediction and data
plt.plot(x,y_pred*15000000000000.)
plt.plot(x,y,'ro')
```

```python
#normalize the data
x_nom=x/max(x)
y_nom=y/max(y)
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmoid,x_nom,y_nom)
```

```python
# plot the final model
x=np.linspace(1960,2015,55)
x=x/max(x)
plt.figure(figsize=(20,10))
y=sigmoid(x,*popt)
plt.plot(x_nom,y_nom,'ro',label="data")
plt.plot(x,y,linewidth=3.0,label="fit")
plt.legend(loc='best')
plt.ylabel("GDP")
plt.xlabel("Year")
plt.show()
```

```python
# split the data
msk=np.random.rand(len(GDP_data)) < 0.8
train_x=x_nom[msk]
test_x=x_nom[~msk]
train_y=y_nom[msk]
test_y=y_nom[~msk]
```

```python
#build the model using train set
popt, pcov  = curve_fit(sigmoid,train_x,train_y)
y_hat=sigmoid(test_x,*popt)
```

```python
print("Mean absolute error: %.2f" % np.mean(np.absolute(y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((y_hat - test_y) ** 2))
from sklearn.metrics import r2_score
print("R2-score: %.2f" % r2_score(y_hat , test_y))
```
